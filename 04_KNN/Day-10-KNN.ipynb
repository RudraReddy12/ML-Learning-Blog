```python
y_train = pd.read_csv('/content/labels_train.csv')
y_train.head()
```

```python
y_test = pd.read_csv('/content/labels_test.csv')
y_test.head()
```

```python
X_test = X_test.iloc[1:,:]
```

```python
X_train = X_train.iloc[1:,:]
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# Train with knn
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler().set_output(transform='pandas')
```

```python
X_train_Transformed = mms.fit_transform(X_train)
X_train_Transformed.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train_Transformed,y_train)
```

```python
X_test_transformed = mms.transform(X_test)
X_test_transformed.head()
```

```python
y_pred = knn.predict(X_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=10,weights = 'distance')
knn.fit(X_train_Transformed,y_train)
y_pred = knn.predict(X_test_transformed)
accuracy_score(y_test,y_pred)
```
