## Segregate train and test data into features and target

```python
x_train=train.drop('target',axis=1)
y_train=train['target']
x_train.shape,y_train.shape
```

```python
x_test=test.drop('target',axis=1)
y_test=test['target']
x_test.shape,y_test.shape
```

```python
# plot some observation
obs=train.sample(1)
obs
```

```python
obs_array = obs.values[:, 1:]
img_array = obs_array.reshape(28, 28)
plt.imshow(img_array, cmap = 'gray')
```

```python
x_train.max()
```

```python
x_train.min()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler().set_output(transform='pandas')
ms
```

```python
x_train_transformed=ms.fit_transform(x_train)
```

```python
x_train_transformed.head()
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(x_train_transformed,y_train)
```

```python
x_test_transformed = ms.transform(x_test)
```

```python
x_test_transformed.head()
```

```python
y_pred=dt.predict(x_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train_transformed,y_train)
```

```python
y_pred=lr.predict(x_test_transformed)
y_pred
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rc=RandomForestClassifier()
```

```python
rc.fit(x_train_transformed,y_train)
```

```python
y_pred=rc.predict(x_test_transformed)
y_pred
```

```python
accuracy_score(y_pred,y_test)
```
