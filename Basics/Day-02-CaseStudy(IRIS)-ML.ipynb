```python
tips['size'].unique() # ordinal categorical
```

```python
def fun1(x):
    if x>4:
        return 'L'
    elif x>2:
        return 'M'
    else:
        return 'S'
```

```python
# size<=2 => small group
# size=3|4 => medium group
# size>=4 => large group
tips['size'].apply(fun1)
```

```python
tips['size_cat']=tips['size'].apply(fun1)
tips.head()
```

```python
# size_cat converting ordinal cat to numerical using ordinal_encoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder().set_output(transform='pandas')
oe
```

```python
oe.fit_transform(tips[['size_cat']])
```

```python
order=['S','M','L']
```

```python
oe1=OrdinalEncoder(categories=[order]).set_output(transform='pandas')
oe1
```

```python
oe1.fit_transform(tips[['size_cat']])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ohe3=OneHotEncoder(sparse_output=False, drop='first')
oe3=OrdinalEncoder(categories=[order])
```

```python
ct=ColumnTransformer([('OHE',ohe3,['sex','smoker','day','time']),
                     ('OE',oe3,['size_cat'])],
                    remainder='passthrough').set_output(transform='pandas')
ct
```

```python
ct.fit_transform()
```
